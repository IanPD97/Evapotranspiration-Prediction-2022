```python
clouds_ = []
weather_ = []
date_ = []
# 12:00
for i in range (1655305200,1656172800,86400):
    url = f'https://api.openweathermap.org/data/3.0/onecall/timemachine?lat={lat}&lon={lon}&dt={i}&appid={api_key}&units={units}'
    req = requests.get(url)
    data = req.json()
    clouds_.append(data['data'][0]['clouds'])
    weather_.append(data['data'][0]['weather'][0]['main'])
    date_.append(datetime.fromtimestamp(data['data'][0]['dt']).strftime('%Y-%m-%d %H:%M:%S'))
```

```python
# 12:00
weathercondition = {'date': date_, 'clouds': clouds_, 'weather_condition': weather_}
df_weathercondition = pd.DataFrame(data = weathercondition)
df_weathercondition.date = pd.to_datetime(df_weathercondition.date,format='%Y-%m-%d')
df_weathercondition = df_weathercondition.set_index('date')

df_weathercondition
```
