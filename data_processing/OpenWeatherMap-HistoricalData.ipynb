```python
RUTA = '/content/drive/MyDrive/Universidad/Evapotranspiration-Prediction-2022/data_processing/'
api_key = '908c5a807c96a89d0f925d87a75cc5b9'
lat = -33.494386
lon = -70.647444
units = 'metric'

timestamp = pd.read_csv('/content/drive/MyDrive/Universidad/Evapotranspiration-Prediction-2022/data_processing/data/Timestamp_.csv')
datatime = timestamp['Timestamp'].values
cont = 0
dt = []
temp = []
pressure = []
humidity = []
clouds = []
wind = []
for i in datatime:
    url = f'https://api.openweathermap.org/data/3.0/onecall/timemachine?lat={lat}&lon={lon}&dt={i}&appid={api_key}&units={units}'
    req = requests.get(url)
    data = req.json()
    clouds.append(data['data'][0]['clouds'])
    temp.append(data['data'][0]['temp'])
    dt.append(data['data'][0]['dt'])
    pressure.append(data['data'][0]['pressure'])
    humidity.append(data['data'][0]['humidity'])
    wind.append(data['data'][0]['wind_speed'])
    cont = cont+1
    if cont==1000:
      df = pd.DataFrame({'Timestamp':datatime})
      df = df['Timestamp'][cont:]
      df.to_csv('/content/drive/MyDrive/Universidad/Evapotranspiration-Prediction-2022/data_processing/data/Timestamp_.csv')
      break
weather_data = pd.DataFrame({'temp':temp,'clouds':clouds,'pressure':pressure,'wind':wind,'humidity':humidity},index=dt)
weather_data.to_csv('/content/drive/MyDrive/Universidad/Evapotranspiration-Prediction-2022/data_processing/data/Weather_data_SM'+datetime.now().strftime(format='%d-%m-%Y')+'.csv')
```
